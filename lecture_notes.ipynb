### Pandas Datastructures
- Series - single dimensional datastructure. Similare to a list or a single column
    - homogeneous(of the same data type)
- DataFrame - multidimensional table
    - heterogeneous (of many datatypes)
    

```python
ser1 = pd.Series([10, 20, 30, 40])
print(ser1)

ser1.iloc[0]
```

```python
# Creating a series from a preexistiong dictionary
d = {
    "b" : 1,
    "a" : 0,
    "c" : 2,
    "d" :100
}
ser2 = pd.Series(d)
```

```python
d1 = {
    "a" : 0.0,
    "b" : 1.0,
    "c" :2.0
}
ser3 = pd.Series(d1)
# ser3 = pd.Series(d1, index = ["apple", 'banana', 'carrot'])
print(ser3)
```

```python
calories = {
    'day1 : 700,'
    'day2' :2500,
    'day3' :1200
}

#use indexes to chose certain values for series
ser4 = pd.Series(calories, index=['day1', 'day3'])
print(ser4)
```

```python
# IF provided a scalar values, value will repeat to fill all index labels
ser5 = pd.Series(5.0, index=['a', 'b', 'c'])
print(ser5)
```

### Data Frames:
- indexes (each record)
- columns (each variable)


```python
fruit_data = {
    'apple_sales': [3, 2, 0, 2, 9, 4, 7, 2],
    'orange_sales': [0, 3, 2, 7, 7, 8, 3, 0 ]
}

#dx = [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007]
idx = pd.RangeIndex(2000, 2008) #Dynamically create a range of indexes

#purchases = pd.DataFrame(fruit_data, index=['Day1', 'Day2', 'Day3', 'Day4', 'Day5', 'Day6', 'Day7', 'Day8'])
purchases = pd.DataFrame(fruit_data, index=idx)
print(purchases)
```

```python
rows, cols = purchases.shape
print(f'Rows: {rows}')
print(f'Cols: {cols}')
```

```python
# To view the top n elements, by default gives 5 elements
print(purchases.head())
```

```python
# To view the bottom n elements, by default 5
print(purchases.tail())
```

### Frequency Analysis - Value counts

```python
# Sample DataFrame
data = {'Category': ['Electronics', 'Clothing', 'Electronics', 'Books', 'Books', 'Clothing']}
cat_df = pd.DataFrame(data)

# .value_counts
cat_df.value_counts()
```

```python
# Create the dataframe
sales_data={"Devices":['Laptop','iPhone','LED','LCD','Smart-Phone','Washing-Machine'],
           'Brand':['Lenovo','Apple','Samsung','Samsung','Samsung','Whirpool'],
           'Sales':[1000,2000,4000,2000,1000,4000],
           'Profit':[500,1000,1000,1500,1000,1500],
           'Pices left':[5000,4000,4000,5000,5000,1000]}

sales_df = pd.DataFrame(sales_data)

Frequency = sales_df ['Brand'].value_counts()
#print(Frequency)

frequent_product = Frequency[Frequency > 1].index[0]

print(Frequency)
print(f'This item appears more than once: {frequent_product}')
```

```python
data = {'Score': [85, 92, 88, 75, None, 90, None, 85]}
scores_df = pd.DataFrame(data)
```

```python
scores_counts = scores_df.value_counts(dropna=False) #dropna, tells function to either include or NOT include Nan, Null, NA values
print('Score', scores_counts)
```

### Rename a column in Pandas
- `.rename()`

```python
tech = {
    'Courses': ["Python", ' DataAnalytics', 'MERN Stack', 'Data Engineering'],
    'Fee': [100, 2000, 250, 300],
    'Duration':['40days', '50days', '3 months', '1 year']
}
tech_df = pd.DataFrame(tech)
```

```python
tech_df = tech_df.rename(columns={'Courses': 'Course Name'})
```
