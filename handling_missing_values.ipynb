### Handling MIssing Values
-   `df.isnull()` - to detect missing values (`.isna()` is the same thing)
-   `df.fillna()` - to fill missing values (imputation)
-   `df.dropna()` - to remove missing values
-   `df.replace()`-  to replace specific values (imputation)

```python
import pandas as pd
import numpy as np
```

```python
employee_list = [['James', 36, 75, 5428000],
               ['Villers', 38, 74, 3428000],
               ['VKole', 31, 70, 8428000],
               ['Smith', 34, 80, 4428000],
               ['Gayle', 40, 100, 4528000],
               ['Adam', 40, np.nan, 4528000],
               ['Rooter', 33, 72, 7028000],
               ['Peterson', 42, 85, 2528000],
               ['lynda', 42, 85, np.nan],
               [np.nan, 42, 85, np.nan],
               ['Jenny', np.nan, 100, 25632],
               ['Kenn', np.nan, 110, 25632],
                ['Aly', np.nan, 90, 25582],
               ['John', 41, 85, 1528000]]

df = pd.DataFrame(employee_list, columns = ['Name', 'Age', 'Weight', 'Salary'])
```

```python
rows, cols = df.shape
print('Rows:', rows)
print('Cols', cols)
```

```python
df.head()
```

```python
# Provide a summarization of missing values per column
df.isnull().sum()
```

```python
# Be wary of missing values, if you have a small dataset, removing values can dramatically alter your analysis
# is missing data ROWS is less than 5% of all data it is safe to remove and will have little effect on analysis

dropped_df = df.dropna()
#when dropping na, you remove the entire row


print(dropped_df)
```

```python
filled_df = df.fillna('None')
filled_df
```

```python
filled_df.isnull().sum()
```

```python
# Only change values of certain columns
specific_cols = df[['Weight', 'Age']].fillna('Pending')

specific_cols
```

```python
df
```

```python
df2 = df.fillna(value={'Weight': 'Pending', 'Age': 'Unknown', 'Salary': '0.0', 'Name': 'Verification Pending'})

df2
```
